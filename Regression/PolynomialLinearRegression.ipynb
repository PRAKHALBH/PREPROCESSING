```python
# Data Preprocessing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values # 1:2 will make it matrix 
y = dataset.iloc[:, 2].values

# from sklearn.model_selection import  train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)

# # Feature Scaling
# ''' from sklearn.preprocessing import StandardScaler
# sc_x = StandardScaler()
# X_train = sc_x.fit_transformn(X_train)
# X_test = sc_x.transform(X_test)'''
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Fitting Polynomial Regression to the dataset 
from sklearn.preprocessing import PolynomialFeatures
# poly_reg obj is going to be transformer tool that will transform x into new matix of features containing x1, x1^2, x1^3....
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
```

```python
# new regression object. Purpose of this is to include this fit that we've made with our poly_reg object
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
# Visualising Linear Regression 
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='green')
plt.title('Linear Plot')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualisign the Polynomial Regression
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)))
plt.title('Polynomial Plot')
plt.xlabel('Postition Salary')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Polynomial Regression (Better plot)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid)), 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)))
plt.title('Polynomial Plot')
plt.xlabel('Postition Salary')
plt.ylabel('Salary')
plt.show()
```
